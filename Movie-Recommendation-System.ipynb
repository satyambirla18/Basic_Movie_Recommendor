```python
# genre
# id
# keywords
# title
# overview
# cast
# crew
movies = movies[['movie_id','genres','title','keywords','overview','cast','crew']]
```

```python
import ast
```

***Function to clean genre column***

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convertcast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies.head()
```

```python
def get_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

***Creating List of String & Joining them***

```python
# movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview'] = movies['overview'].apply(lambda x: ' '.join(x))
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

***Removing Spaces(" ")***

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies = movies[['movie_id','title','tags']]
```

```python
new_movies
```

***Creating & joining Strings into Lists***

```python
# new_movies['tags'] = movies['tags'].apply(lambda x:" ",join(x))
new_movies['tags'] = new_movies['tags'].apply(lambda x: ' '.join(x))
new_movies['tags'] = new_movies['tags'].apply(lambda x: x.split())
```

```python
new_movies
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    if isinstance(text, list):  # Check if the input is a list
        y = []
        for i in text:
            y.append(ps.stem(i))
        return y
    else:  # If the input is not a list, assume it's a string
        y = []
        for i in text.split():
            y.append(ps.stem(i))
        return " ".join(y)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
# new_movies['tags'] = new_movies['tags'].apply(lambda x:x.lower())
# Assuming 'tags' column contains lists of strings
new_movies['tags'] = new_movies['tags'].apply(lambda tags: [tag.lower() for tag in tags])
```

```python
new_movies
```

***Vectorization of Text***

```python
from sklearn.feature_extraction.text import CountVectorizer
# Custom preprocessor function
def custom_preprocessor(text):
    if isinstance(text, list):
        return ' '.join(text)
    else:
        return text

# Update CountVectorizer to use custom preprocessor and tokenizer
cv = CountVectorizer(max_features=5000, preprocessor=custom_preprocessor)  

# Fit the CountVectorizer to your data
try:
    vectors = cv.fit_transform(new_movies['tags']).toarray()
    feature_names = cv.get_feature_names_out()
except ValueError as e:
    print("ValueError occurred:", e)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_idx = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity_matrix[movie_idx]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
